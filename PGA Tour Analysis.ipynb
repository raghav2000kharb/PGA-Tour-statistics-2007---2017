```python
### Reading pgatour.csv data into the jupyter notebook as df
```

```python

import pandas as pd 
df = pd.read_csv('/home/raghav/Desktop/College/Data Science/pgatour_raw.csv', encoding="Latin-1")
df
```

### Cleaning MONEY & POINTS column 

```python

df['MONEY'] = df['MONEY'].str.replace(',', '')
df['MONEY'] = df['MONEY'].str.replace('$', '')
df['POINTS'] = df['POINTS'].str.replace(',', '')
df.info()
```

### Renaming FWY_% and GIR_% columns for performing agg functions later in the analysis

```python
df = df.rename(columns={"FWY_%": "FWY_PCT", "GIR_%": "GIR_PCT"})
df.info()
```

### Replacing the blank and null values from the data set and converting POINTS and MONEY to numeric data type
 

```python
df.fillna(0, inplace=True)
df.replace('', '0', inplace=True)
df[['POINTS']] = df[['POINTS']].apply(pd.to_numeric)
df[['MONEY']] = df[['MONEY']].apply(pd.to_numeric)
df.info()
df
```

### Average scoring of all the players year-wise

```python
year_score_avg=df.groupby(by='Year')['SCORING'].mean().reset_index()
print('Scoring Averages From 2007 - 2017 on PGA Tour\n')
print(year_score_avg)
```

Scoring average of the tour has remained similair during the span of 10 years

### Lowest average score from 2007 -2017

```python
year_score_low=(year_score_avg[year_score_avg.SCORING == year_score_avg.SCORING.min()]) 
print('Lowest Average Score\n')
print(year_score_low)
```

### Highest average score from 2007-2017

```python
year_score_max=(year_score_avg[year_score_avg.SCORING == year_score_avg.SCORING.max()]) 
print('Highest Average Score\n')
print(year_score_max)
```

 ### Finding mean of driving distance, fairway hit%, scoring, gir%, strokes gained, money, points of players from 2007 - 2017

```python
mean_df=df.groupby('NAME').agg({'SCORING': 'mean','DRIVE_DISTANCE': 'mean', 'FWY_PCT': 'mean', 'GIR_PCT': 'mean', 'SG_T': 'mean','SG_P': 'mean','SG_TTG': 'mean','MONEY':'mean', 'POINTS': 'mean' }).reset_index()
print('Average Data of players from 2007 - 2017\n')
mean_df
```

### Player with longest distance off the tee from 2007-2017

```python
max_dd=(mean_df[mean_df.DRIVE_DISTANCE == mean_df.DRIVE_DISTANCE.max()])
print('Longest Average Driving Distance \n')
max_dd
```

### Player with shortest distance off the tee from 2007-2017

```python
low_dd=(mean_df[mean_df.DRIVE_DISTANCE == mean_df.DRIVE_DISTANCE.min()])
print('Shortest Average Driving Distance\n')
low_dd
```

### Player with highest Fairway hit percetage of the tee from 2007-2017

```python
max_fwy=(mean_df[mean_df.FWY_PCT == mean_df.FWY_PCT.max()])
print('Highest Fairways Hit %\n')
max_fwy
```

### Player with lowest Fairway hit percetage of the tee from 2007-2017

```python
low_fwy=(mean_df[mean_df.FWY_PCT == mean_df.FWY_PCT.min()])
print('Lowest Fairways Hit%\n')
low_fwy
```

### Finding avergare distance off the tee and average fairway hit percentage of the field from 2007-2017

```python
avg_dd =mean_df['DRIVE_DISTANCE'].mean()
avg_fwy= mean_df['FWY_PCT'].mean()
print("Average Driving Distance of the tee from 2007-2017")
print(avg_dd)
print("\nAverage Fairway Hit % from 2007-2017")
print(avg_fwy)

```

### Year wise average of Scoring, Drive_Distance, Fairway%, GIR%, Strokes Gained-(Total, Putting, Tee To Green), Money and Points

```python
mean_year_df=df.groupby('Year').agg({'SCORING': 'mean','DRIVE_DISTANCE': 'mean', 'FWY_PCT': 'mean', 'GIR_PCT': 'mean', 'SG_T': 'mean','SG_P': 'mean','SG_TTG': 'mean','MONEY':'mean', 'POINTS': 'mean' }).reset_index()
mean_year_df
```
